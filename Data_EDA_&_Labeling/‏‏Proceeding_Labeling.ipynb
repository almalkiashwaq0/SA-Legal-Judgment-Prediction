```python
import re
def clean_text(text):
    arabic_pattern = r'[^\u0600-\u06FF\s]'
    text = re.sub(arabic_pattern, ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df['Judgment'] = df['Judgment'].apply(clean_text)
```
