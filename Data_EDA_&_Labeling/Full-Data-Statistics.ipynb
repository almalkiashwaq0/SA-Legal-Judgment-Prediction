```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
```

```python
df = pd.read_csv("Dataset.csv")
df
```

```python
df['Word_Count'] = df['Facts'].apply(lambda x: len(str(x).split()))
```

```python
df['Word_Count'].min()
```

```python
df['Word_Count'].max()
```

```python
fig = go.Figure()

fig.add_trace(go.Box(
    y=df['Word_Count'],
    name='Word Count',
    marker=dict(color='#0072B2'),
    boxmean=True  
))


fig.update_layout(
    title='Word Count Distribution in Facts Column',
    yaxis_title='Word Count',
    height=400,
    width=600
)
fig.show()
```

```python
city_counts = df['City'].value_counts().reset_index()
city_counts.columns = ['City', 'Count']

fig = px.bar(
    city_counts,
    x='City',
    y='Count',
    color_discrete_sequence=['#0072B2'],  
    labels={'City': 'City', 'Count': 'Count'},
    title='City Distribution in Dataset'
)
fig.show()
```

```python
df = df[df['City'].str.match(r'^[\u0600-\u06FF]+$', na=False)]

city_counts = df['City'].value_counts().reset_index()
city_counts.columns = ['City', 'Count']

fig = px.bar(
    city_counts,
    x='City',
    y='Count',
    color_discrete_sequence=['#0072B2'],  
    labels={'City': 'City', 'Count': 'Count'},
    title='City Distribution in Dataset (Arabic Only)'
)
fig.show()
```

```python
print("Unique values in 'Type' column:", df['Type'].unique())

type_mapping = {
    ' ': 'Proceedings of the case',
    '': 'Renting',
    '': 'Selling',
}

df['Type'] = df['Type'].replace(type_mapping)

fig = px.bar(df['Type'].value_counts(), 
             x=df['Type'].value_counts().index, 
             y=df['Type'].value_counts().values, 
             labels={'x': 'Type', 'y': 'Count'}, 
             title='Distribution of Case Types')
fig.show()
```

```python
type_mapping = {
    ' ': 'Proceedings of the case',
    '': 'Renting',
    '': 'Selling'
}
df['Type'] = df['Type'].replace(type_mapping)

label_mapping = {
    '': 'Rejected',
    '': 'Accepted'
}
df['Label'] = df['Label'].replace(label_mapping)

custom_colors = {
    'Rejected': '#E69F00',
    'Accepted': '#0072B2'
}

fig = px.histogram(
    df,
    x='Type',
    color='Label',
    title='Distribution of Labels in Each Case Type',
    labels={'Type': 'Case Type', 'count': 'Count'},
    barmode='group',
    color_discrete_map=custom_colors
)

fig.update_layout(
    xaxis_title='Case Type',
    yaxis_title='Count',
    legend_title_text='Label'
)

fig.show()
```

```python
type_translation = {
    '': 'Renting',
    '': 'Selling',
    ' ': 'Proceedings'
}

df['Type'] = df['Type'].map(type_translation)

summary = df.groupby(['Type', 'Encoded_Label']).size().unstack(fill_value=0)
summary.columns = ['Rejected', 'Accepted']


summary['Total'] = summary['Rejected'] + summary['Accepted']
print(summary)
```
