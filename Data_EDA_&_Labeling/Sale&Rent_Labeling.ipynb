```python
def clean_text(text):
    text = re.sub(r'<[^>]+>', '', text)
    text = re.sub(r'[\n\r\t\f\v\u00A0\u200B]', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

columns_to_clean = ['Judgment', 'Facts', 'Reasons']

for column in columns_to_clean:
    if column in df.columns:  
        df[column] = df[column].apply(clean_text)
```

```python
def add_space_around_punctuation(text):
    return re.sub(r'(\w)([^\w\s])', r'\1 \2', re.sub(r'([^\w\s])(\w)', r'\1 \2', text))

columns_to_clean = ['Judgment', 'Facts', 'Reasons']

for column in columns_to_clean:
    if column in df.columns:  
        df[column] = df[column].apply(add_space_around_punctuation)
```
